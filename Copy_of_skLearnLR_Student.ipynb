### ML Basic Concept  with Python

```
from sklearn import linear_model
from sklearn.metrics import mean_squared_error, r2_score

# Create linear regression object
regr = linear_model.LinearRegression()
regr.fit(x, y)
y_pred = regr.predict(x)

```



````
from sklearn import linear_model
from sklearn.metrics import mean_squared_error, r2_score
````

````
x.shape #1มิติ
````

````
print(x.reshape(12,1)) #ไลบรารีรับ 2 มิติ จึงต้องเปลี่ยนเป็นเเนวตั้ง
````

#Reshapeข้อมูลให้รันผ่าน

````
x_train=x.reshape(-1,1) #-1เป็นdummy
````

````
# Create linear regression object

reMod= linear_model.LinearRegression()              #declare object
reMod.fit(x_train, y)                                #Train
````

````
reMod.predict([[50]]) #ทายค่า
````

### Performance 
```
print('Mean squared error: %.2f' % mean_squared_error(y, y_pred))
print('Coefficient of determination: %.2f' % r2_score(y, y_pred))
```



````
y_pred=reMod.predict(x_train)
y_pred
````

````
plt.scatter(x,y,c='green') #scatter เปลี่ยนสีไม่เหมือนกัน.
plt.plot(x,y_pred,'r')
````
